```python
%matplotlib inline

import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import image as image
import easygui
```

```python

# 1.Open any image
```

```python
I = cv2.imread("image.jpg")
```

```python

# 2.Build a kernel for sharpening an image
```

```python
k = np.ones((5,5),np.uint8)
```

```python

# 3.Apply this filter to the image
```

```python
F = cv2.filter2D(I,ddepth=-1, kernel=k)
plt.imshow(F)
plt.show()
```

```python
cv2.imshow("Image",F)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imgCanny = cv2.Canny(I,100,200)
plt.imshow(imgCanny)
plt.show()

cv2.imshow("Canny",imgCanny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imgDialation = cv2.dilate(imgCanny, k, iterations=1)
plt.imshow(imgDialation)
plt.show()

cv2.imshow("Dialation",imgDialation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
