```python
%matplotlib inline

import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import image as image
import easygui 
```

```python
f = easygui.fileopenbox(filetypes=["*.jpg", "*.jpeg", "*.png"])
I = cv2.imread(f) # image read (user input)
```

```python
I = cv2.imread('image.jpg') # direct input
```

```python
img_YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
cv2.imshow("original", I)
cv2.imshow("YUV",img_YUV)
cv2.waitKey()
cv2.destroyAllWindows() #openCV convert image to YUV
```

```python
YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
plt.imshow(I)
plt.show()

plt.imshow(YUV)
plt.show() # MatPlotLib convert image to YUV
```

```python
G = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
cv2.imshow("Original", I)
cv2.imshow("Gray Scale",G)
key = cv2.waitKey(0)
cv2.destroyAllWindows() # openCV show grayscale
```

```python
G = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.imshow(I)
plt.show()

plt.imshow(G)
plt.show() # MatPlotLib show grayscale
```
