```python

# 3. Capture the user's click on the image
```

```python
cv2.imshow("image",I)
def draw(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("Point:=",x,y)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.destroyAllWindows()
        
cv2.namedWindow("image")
cv2.setMouseCallback("image",draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# 4. Draw a 201 x 201, 5-pixel thick red square around this location
```

```python
cv2.imshow("image",I)
def draw(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("Point:=",x,y)
        x1 = x - 100
        x2 = x + 101
        y1 = y - 100 
        y2 = y + 101
        cv2.rectangle(img = I, pt1 = (x1,y1), pt2 = (x2,y2), 
                      color = (0,0,255), thickness = 5)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.imshow("image",I)
        
cv2.namedWindow("image")
cv2.setMouseCallback("image",draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# 5. Convert the pixels within the square to YUV
```

```python
cv2.imshow("image",I)
def draw(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        print("Point:=",x,y)
        x1 = x - 100
        x2 = x + 101
        y1 = y - 100 
        y2 = y + 101
        cv2.rectangle(img = I, pt1 = (x1,y1), pt2 = (x2,y2), 
                      color = (0,0,255), thickness = 5)
        img_select = I[y1:y2, x1:x2]
        img_YUV = cv2.cvtColor(img_select, cv2.COLOR_BGR2YUV)
        I[y1:y2, x1:x2] = img_YUV
        cv2.imshow("image",I) # update image
        
cv2.namedWindow("image")
cv2.setMouseCallback("image",draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
