```python
%matplotlib inline

import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import image as image
import easygui 
```

```python

# 1. Read in any image
```

```python
f = easygui.fileopenbox(filetypes=["*.jpg", "*.jpeg", "*.png"])
I = cv2.imread(f) # image read (user input)
```

```python
I = cv2.imread('image.png') # direct input
```

```python

# 2. Convert this image to YUV
```

```python
img_YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
cv2.imshow("original", I)
cv2.imshow("YUV",img_YUV)
cv2.waitKey()
cv2.destroyAllWindows() #openCV convert image to YUV
```

```python
YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
plt.imshow(I)
plt.show()

plt.imshow(YUV)
plt.show() # MatPlotLib convert image to YUV
```

```python

# 3. Extract the Y channel (luminance)
```

```python
y, u, v = cv2.split(img_YUV)
print(y)
cv2.imshow("Luminance", y)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python

# 4. Convert the original image to grayscale（intensity)
```

```python
G = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
cv2.imshow("Original", I)
cv2.imshow("Gray Scale",G)
key = cv2.waitKey(0)
cv2.destroyAllWindows() # openCV show grayscale
```

```python
G = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
plt.imshow(I)
plt.show()

plt.imshow(G)
plt.show()

plt.imshow(G,cmap='gray')
plt.show() # MatPlotLib show grayscale
```

```python

# 5. Show the two on the screen using OpenCV's imshow function
```

```python
mix = np.hstack((I,img_YUV))
cv2.imshow("Result", mix)
cv2.waitKey(0)   
cv2.destroyAllWindows()
```

```python

# 6. Show the luminance on the screen side by side with 
# the intensity using MatPlotLib's function
```

```python
YUV = np.hstack((y,u))
YUV2 = np.hstack((YUV,v))

plt.imshow(YUV2)
plt.show() 
```
