```python
%matplotlib inline

import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import image as image
import easygui
```

```python

# 1.Open image "Orange.png" and "Water.jpg"
```

```python
Orange = cv2.imread("Orange.png")
Water = cv2.imread("Water.jpg")
```

```python
Orange1 = Orange[:,:,[2,1,0]]
Water1 = Water[:,:,[2,1,0]]
plt.imshow(Orange1)
plt.show()

plt.imshow(Water1)
plt.show()
```

```python

# 2.Scale each image to 50% by multiplying each by 0.5
```

```python
h1,w1,d1 = Orange1.shape
h2,w2,d2 = Water1.shape

dstw1 = int(w1*0.5)
dstw2 = int(w2*0.5)
dsth1 = int(h1*0.5)
dsth2 = int(h2*0.5)

SO = cv2.resize(Orange1,dsize=(dstw1,dsth1))
SW = cv2.resize(Water1,dsize=(dstw2,dsth2))

plt.imshow(SO)
plt.show()

plt.imshow(SW)
plt.show()
```

```python

# 3.Add the two scaled images to get a composite image
```

```python
A = cv2.add(SO, SW)

plt.imshow(A)
plt.show()
```

```python

# 4.Adjust the scaling to give a nicer output
```

```python
M1 = cv2.multiply(SO,SW,scale=0.01)

plt.imshow(M1)
plt.show()
```

```python
AW = cv2.addWeighted(SO,0.7,SW,0.5,0.0)

plt.imshow(AW)
plt.show()
```
